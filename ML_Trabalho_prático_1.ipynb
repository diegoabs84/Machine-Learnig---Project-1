```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 
from sklearn import preprocessing
```

```python
#lendo os dados das famílias
datafam = pd.read_csv("./data_familia_sergipe.csv")
datafam.head(10)
```

```python
#lendo os dados das pessoas
datap = pd.read_csv("./data_pessoas_sergipe.csv")
datap.head(10)
```

Em virtude do manual do pesquisador nos informar que o bolsa família leva em consideração um certo intervalo de idade para estabelecer o montante a ser repassado por pessoa àquela família, pode-se transformar o critério "idade", do dataset referente a pessoas, em uma medida qualitativa.
Vamos utilizar a seguinte classificação de idades: Criança (0-15 anos), adolescente (16-17 anos), adulto (18-59 anos) e idoso (acima de 60 anos).

```python
datap['idade'] = pd.cut(datap['idade'],bins=[0,16,19,60,float("inf")],right=False,labels=['criança','adolescente','adulto','idoso'])
datap.head(20)
```

Pode-se observar que, embora alguns tipos sejam especificados como quantitativos, na verdade eles são qualitativos. Deve-se transformá-los para qualitativos. Iniciando pelo dataset família:



```python
for col in ['cd_ibge','dat_alteracao_fam', 'dat_cadastramento_fam', 'cod_local_domic_fam', 'cod_especie_domic_fam', 'cod_material_piso_fam', 'cod_material_domic_fam', 'cod_escoa_sanitario_domic_fam', 'cod_destino_lixo_domic_fam', 'cod_destino_lixo_domic_fam', 'cod_banheiro_domic_fam', 'cod_iluminacao_domic_fam', 'cod_calcamento_domic_fam', 'dat_atualizacao_familia', 'cod_agua_canalizada_fam', 'cod_abaste_agua_domic_fam', 'cod_familia_indigena_fam', 'nom_centro_assist_fam', 'ind_familia_quilombola_fam', 'cod_eas_fam', 'ind_familia_quilombola_fam', 'cod_centro_assist_fam', 'ind_parc_mds_fam', 'id_familia','estrato', 'classf', 'marc_pbf','nom_estab_assist_saude_fam']:
  datafam[col] = datafam[col].astype('category')

datafam.dtypes
```

O mesmo sendo feito para o dataset de pessoas:


```python
for col in ['cd_ibge', 'estrato', 'classf', 'id_familia', 'id_pessoa', 'cod_sexo_pessoa', 'cod_parentesco_rf_pessoa', 'cod_raca_cor_pessoa', 'cod_local_nascimento_pessoa', 'cod_certidao_registrada_pessoa', 'cod_deficiencia_memb', 'cod_sabe_ler_escrever_memb', 'ind_frequenta_escola_memb', 'cod_escola_local_memb', 'cod_curso_frequenta_memb', 'cod_ano_serie_frequenta_memb', 'cod_curso_frequentou_pessoa_memb', 'cod_ano_serie_frequentou_memb', 'cod_concluiu_frequentou_memb', 'cod_trabalhou_memb', 'cod_afastado_trab_memb', 'cod_agricultura_trab_memb', 'cod_principal_trab_memb', 'cod_trabalho_12_meses_memb']:
  datap[col] = datap[col].astype('category')
datap.dtypes
```

Outro ponto a se observar é que as colunas referentes ao pesos das pessoas e das famílias estão em uma escala muito superior às demais. Deve-se normalizar estas escalas:


O primeiro passo é converter a coluna que será normalizada para float.


```python
float_array = datafam['peso.fam'].values
float_array1 = datap['peso.fam'].values
float_array2 = datap['peso.pes'].values
```

O segundo passo é criar o objeto min/max do preprocessing.


```python
min_max_scaler = preprocessing.MinMaxScaler()
scaled_array = min_max_scaler.fit_transform(float_array[:, np.newaxis])
scaled_array1 = min_max_scaler.fit_transform(float_array1[:, np.newaxis])
scaled_array2 = min_max_scaler.fit_transform(float_array2[:, np.newaxis])
```

O terceiro passo é substituir os valores na base.

```python
datafam['peso.fam'] = scaled_array
datap['peso.fam'] = scaled_array1
datap['peso.pes'] = scaled_array2
datafam.head(10)
```
