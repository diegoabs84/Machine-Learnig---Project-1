```python
#Vamos procurar normalizar alguns campos que estão com a escala muito grande no dataset
# O primeiro passo é converter a coluna que será normalizada para float. 
float_array = datafam['peso.fam'].values
float_array1 = datafam['cod_centro_assist_fam'].values
```

```python
# O segundo passo é criar o objeto min/max do preprocessing

min_max_scaler = preprocessing.MinMaxScaler()
scaled_array = min_max_scaler.fit_transform(float_array[:, np.newaxis])
scaled_array1 = min_max_scaler.fit_transform(float_array1[:, np.newaxis])
```

```python
# O terceiro passo é substituir os valores na base
datafam['peso.fam'] = scaled_array
datafam['cod_centro_assist_fam'] = scaled_array1
datafam.head(10)
```

```python
#Agora, normalizando os valores do dataset pessoas
float_array2 = datap['peso.fam'].values
float_array3 = datap['peso.pes'].values
```

```python
# O segundo passo é criar o objeto min/max do preprocessing

min_max_scaler = preprocessing.MinMaxScaler()
scaled_array2 = min_max_scaler.fit_transform(float_array2[:, np.newaxis])
scaled_array3 = min_max_scaler.fit_transform(float_array3[:, np.newaxis])
```

```python
# O terceiro passo é substituir os valores na base
datap['peso.fam'] = scaled_array2
datap['peso.pes'] = scaled_array3
datap.head(10)
```
